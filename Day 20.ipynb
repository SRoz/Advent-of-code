```python
import numpy as np
import string
```

```python
LETTERS = [l for l in string.ascii_uppercase]
```

```python
def to_array(string):
    lines = string.split("\n")
    return np.array([[l for l in t] for t in lines])
```

```python
def get_letter(letter, array):
    matches = np.where(array==letter)
    return [(mi, mj) for mi, mj in zip(matches[0], matches[1])]
```

```python
def try_dir(pos, array):
    to_ret = False
    try:
        if (array[pos[0]-1, pos[1]] == '.') and (array[pos[0]+1,pos[1]] in LETTERS):
            return (array[pos[0],pos[1]]+array[pos[0]+1,pos[1]], (pos[0]-1, pos[1]))
    except:
        pass
    try:
        if (array[pos[0]+2, pos[1]] == '.') and (array[pos[0]+1,pos[1]] in LETTERS):
            return (array[pos[0],pos[1]]+array[pos[0]+1,pos[1]], (pos[0]+2, pos[1]))
    except:
        pass
    try:
        if (array[pos[0], pos[1]-1] == '.') and (array[pos[0],pos[1]+1] in LETTERS):
            return (array[pos[0],pos[1]]+array[pos[0],pos[1]+1], (pos[0], pos[1]-1))
    except:
        pass
    try:
        if (array[pos[0], pos[1]+2] == '.') and (array[pos[0],pos[1]+1] in LETTERS):
            return (array[pos[0],pos[1]]+array[pos[0],pos[1]+1], (pos[0], pos[1]+2))
    except:
        pass
    return False
```

```python
def get_gates(array):
    gates = []
    for L in LETTERS:
        positions = get_letter(L, array)

        if len(positions)>0:
            for p in positions:
                r = try_dir(p, array)
                if r:
                    gates.append(r)
    return gates
```

```python
def travel(array, pos, visited, connections, solutions):
    solutions=solutions.copy()
    visited=visited.copy()
    if not array[pos[0], pos[1]] == '.':
        return False
    if pos in visited:
        return False
    if (pos==connections['start']) and (len(visited)>0):
        return False
    if pos==connections['end']:
        solutions.append(len(visited))
        print(solutions)
        return False
    
    if pos in connections.keys():
        visited.append(pos)
        pos = connections[pos]
    
    visited.append(pos)
    
    if (travel(array, (pos[0]+1, pos[1]), visited, connections, solutions) or
        travel(array, (pos[0]-1, pos[1]), visited, connections, solutions) or
        travel(array, (pos[0], pos[1]+1), visited, connections, solutions) or
        travel(array, (pos[0], pos[1]-1), visited, connections, solutions)):
        return True
    return False
        
```

```python
def try_dir(pos, array):
    to_ret = False
    try:
        if (array[pos[0]-1, pos[1]] == '.') and (array[pos[0]+1,pos[1]] in LETTERS):
            to_ret = (array[pos[0],pos[1]]+array[pos[0]+1,pos[1]], (pos[0]-1, pos[1]))
    except:
        pass
    try:
        if (array[pos[0]+2, pos[1]] == '.') and (array[pos[0]+1,pos[1]] in LETTERS):
            to_ret = (array[pos[0],pos[1]]+array[pos[0]+1,pos[1]], (pos[0]+2, pos[1]))
    except:
        pass
    try:
        if (array[pos[0], pos[1]-1] == '.') and (array[pos[0],pos[1]+1] in LETTERS):
            to_ret = (array[pos[0],pos[1]]+array[pos[0],pos[1]+1], (pos[0], pos[1]-1))
    except:
        pass
    try:
        if (array[pos[0], pos[1]+2] == '.') and (array[pos[0],pos[1]+1] in LETTERS):
            to_ret = (array[pos[0],pos[1]]+array[pos[0],pos[1]+1], (pos[0], pos[1]+2))
    except:
        pass
    
    outer = (pos[0] == 0) or (pos[1]==0) or (pos[0]==array.shape[0]-2) or (pos[1]==array.shape[1]-2)
    
    if to_ret:
        to_ret = (to_ret[0], to_ret[1], outer)
    return to_ret
```

```python
def get_gates(array):
    gates = []
    for L in LETTERS:
        positions = get_letter(L, array)

        if len(positions)>0:
            for p in positions:
                r = try_dir(p, array)
                if r:
                    gates.append(r)
    return gates
```

```python
def travel(array, pos, level, visited, connections, inout, solutions, in_gates):
    global best_answer
    
    solutions=solutions.copy()
    visited=visited.copy()
    in_gates = in_gates.copy()
    
    if len(visited) > best_answer:
        return False
    
    if level > 25:
        return False
    
    if not array[pos[0], pos[1]] == '.':
        return False
    if (pos, level) in visited:
        return False
    if (pos==connections['start']) and (len(visited)>0):
        return False
    if (pos==connections['end']) & (level==0):
        best_answer = len(visited)
        print(best_answer)
        return False
    
    if pos in connections.keys():
        visited.append((pos, level))
        
        #if pos in in_gates:
            #print(level)
            #return False
        
        if inout[pos]:
            if level ==0:
                return False
            level -= 1
            #print(level)
            if connections[pos] in in_gates:
                in_gates.remove(connections[pos])
            
        else:
            level += 1
            in_gates.append(pos)
            
        
        pos = connections[pos]
    
    visited.append((pos, level))
    
    if (travel(array, (pos[0]+1, pos[1]), level, visited, connections, inout, solutions, in_gates) or
        travel(array, (pos[0]-1, pos[1]), level, visited, connections, inout, solutions, in_gates) or
        travel(array, (pos[0], pos[1]+1), level, visited, connections, inout, solutions, in_gates) or
        travel(array, (pos[0], pos[1]-1), level, visited, connections, inout, solutions, in_gates)):
        return True
    return False
        
```
