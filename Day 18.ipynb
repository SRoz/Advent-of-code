```python
import numpy as np
import string
from functools import lru_cache
```

```python
letters = [l for l in string.ascii_lowercase]
LETTERS = [l for l in string.ascii_uppercase]
```

```python
def to_array(string):
    lines = string.split("\n")
    return np.array([[l for l in t] for t in lines])
```

```python
def find_idx(grid, val):
    array = to_array(grid)
    idx = np.where(array==val)
    if len(idx[0])>0:
        return [idx[0][0], idx[1][0]]
    else:
        return False
```

```python
def pathfinder(grid, pos, end, gates=[], route=[]):
    gates = gates.copy()
    route = route.copy()
    global best_path, best_gates
    if pos in route:
        return False
    route += [pos]
    
    if pos==end:
        best_path = len(route)-1
        best_gates = gates.copy()
        return False
    if (len(route) > best_path) & (best_path!=-1):
        return False
    
    val = grid[pos[0], pos[1]]
    if val == '#':
        return False
    if val in LETTERS:
        gates += [val]
    
    if (pathfinder(grid, [pos[0] + 1, pos[1]], end, gates, route) or
        pathfinder(grid, [pos[0] - 1, pos[1]], end, gates, route) or
        pathfinder(grid, [pos[0], pos[1] + 1], end, gates, route) or
        pathfinder(grid, [pos[0], pos[1] - 1], end, gates, route)):
        return True
    return False
        
```

```python
@lru_cache(maxsize=None)
def shortest_path(grid, key1, key2):
    """Return pathlength and keys required"""
    global best_path, best_gates
    
    print(f"Finding {key1}, {key2}")
    
    best_path = -1
    array = to_array(grid)
    
    pathfinder(array,
               find_idx(grid, key1),
               find_idx(grid, key2)
              )
    return best_path, best_gates.copy()
```

```python
def all_keys(grid):
    return [l for l in letters if find_idx(grid, l)]
```

```python
def add_level(grid, paths=[(["@"], 0)], keys=letters):
    to_ret = []
    paths = paths.copy()
    for p in paths:
        for k in keys:
            if k in p[0]:
                continue
                
            l, r = p[0][-1], k
            l, r = (l, r) if l < r else (r, l)
            path_len, gates = shortest_path(grid, l, r)
            if set([g.lower() for g in gates]).issubset(set(p[0])):
                to_ret.append((p[0]+[k], p[1]+path_len))
    return to_ret
```

```python
def reduce_paths(paths):
    to_ret = []
    p0s = set()
    sorted_paths = sorted(paths, key = lambda p: p[1])
    sorted_paths = [(sorted(p[0][:-1]) + [p[0][-1]], p[1]) for p in sorted_paths]
    for p in sorted_paths:
        if tuple(p[0]) not in p0s:
            to_ret.append(p)
            p0s.add(tuple(p[0]))
    return to_ret
```

```python
def q1(grid):
    keys = all_keys(grid)
    paths = add_level(grid, keys=keys)
    
    for _ in range(len(keys)-1):
        print(f"{_+1}/{len(keys)-1}")
        print(len(paths))
        paths = add_level(grid, paths, keys)
        paths = reduce_paths(paths)

    minpath = min([p[1] for p in paths])
    return [p for p in paths if p[1]==minpath]
```

```python
def find_idx(grid, val):
    array = to_array(grid)
    idx = np.where(array==val)
    if len(idx[0])==1:
        return [idx[0][0], idx[1][0]]
    elif len(idx[0]>1):
        return [[idx[0][i], idx[1][i]] for i in range(4)]
    else:
        return False
```

```python
@lru_cache(maxsize=None)
def shortest_path(grid, key1, key2):
    """Return pathlength and keys required"""
    global best_path, best_gates
    
    print(f"Finding {key1}, {key2}")
    
    best_path = -1
    array = to_array(grid)
    
    if key1!="@":
        pathfinder(array,
                   find_idx(grid, key1),
                   find_idx(grid, key2)
                  )
        return best_path, best_gates.copy()
    else:
        for at in find_idx(grid, "@"):
            best_path = -1
            pathfinder(array,
                      at,
                      find_idx(grid, key2)
                      )
            if best_path != -1:
                return best_path, best_gates.copy()
```

```python
def get_quartiles(grid):
    gs = to_array(grid).shape
    gs = [gs[0]/2, gs[1]/2]
    quartiles = dict()
    for l in letters:
        loc = find_idx(grid, l)
        if loc:
            if (loc[0] < gs[0]) & (loc[1] < gs[1]):
                quartiles[l] = 'Q1'
            elif (loc[0] < gs[0]) & (loc[1] > gs[1]):
                quartiles[l] = 'Q2'
            elif (loc[0] > gs[0]) & (loc[1] > gs[1]):
                quartiles[l] = 'Q3'
            else:
                quartiles[l] = 'Q4'
    return quartiles
```

```python
def add_level(grid, quartiles, paths=[(["@"], 0)], keys=letters):
    to_ret = []
    paths = paths.copy()
    for p in paths:
        for k in keys:
            if k in p[0]:
                continue
                
            for i in range(len(p[0])-1, -1, -1):
                l = p[0][i]
                if l=='@':
                    break
                if quartiles[l] == quartiles[k]:
                    break
            l, r = (l, k) if l < k else (k, l)
            path_len, gates = shortest_path(grid, l, r)
            
            if set([g.lower() for g in gates]).issubset(set(p[0])):
                to_ret.append((p[0]+[k], p[1]+path_len))
    return to_ret
```

```python
def sort_p(paths, quartiles):
    lasts = []
    firsts = []
    paths = [p for p in paths if p != "@"].copy()
    for q in ['Q1', 'Q2', 'Q3', 'Q4']:
        l = [p for p in paths if quartiles[p]==q]
        if len(l)>0:
            lasts.append(l.pop(-1))
            firsts += l
    to_ret = ["@"] + sorted(firsts) + sorted(lasts)
    return to_ret
        
```

```python
def reduce_paths(paths, quartiles):
    to_ret=[]
    p0s = set()
    sorted_paths = sorted(paths, key = lambda p: p[1])
    sorted_paths = [(sort_p(p[0], quartiles), p[1]) for p in sorted_paths]
    for p in sorted_paths:
        if tuple(p[0]) not in p0s:
            to_ret.append(p)
            p0s.add(tuple(p[0]))
    return to_ret
```
