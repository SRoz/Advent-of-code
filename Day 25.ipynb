```python
with open('data/day25.txt') as fn:
    program = fn.readlines()
    program = [int(i) for i in program[0].split(',')]
```

```python
class ASCII(IntcodeComputer):
    def run_code(self, inp):
        instruction = []
        for line in inp:
            instruction += [ord(i) for i in line] + [10]
        self.run(instruction)

    def draw_ascii(self):
        print("".join([chr(x) for x in self.output]))
        
```

```python
items = ['coin',
        'candy cane',
        'tambourine',
        'fuel cell',
        'mutex']

all_combs = []
for r in range(1,7):
    all_combs += [c for c in combinations(items, r=r)]
```
