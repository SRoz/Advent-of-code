```python
def q1(x, y):
    feed_x = []
    n_beam_x = 0

    i=0
    j=y
    while True:
        drone = IntcodeComputer(program)
        next_square = drone.run([i, j])
        if (n_beam_x>0) and (next_square!=1):
            break
        feed_x.append(next_square)
        if (next_square==1) & (i>=x):
            n_beam_x+=1

        i+=1     

    feed_y = []
    n_beam_y = 0

    i=x
    j=0
    while True:
        drone = IntcodeComputer(program)
        next_square = drone.run([i,j])
        
        if (n_beam_y>0) and (next_square!=1):
            break
        feed_y.append(next_square)
        if (next_square==1) & (j>=y):
            n_beam_y+=1

        j+=1
    print(n_beam_x)
    print(n_beam_y)
        
    return feed_x, feed_y
```
